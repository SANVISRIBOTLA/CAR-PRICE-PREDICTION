```python
X = car[["symboling", "wheelbase", "carlength", "carwidth", "carheight", "curbweight", "enginesize", "boreratio", "stroke", "compressionratio", "horsepower", "peakrpm", "citympg", "highwaympg"]]
```

```python
y = car["price"]
```

# Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Training a Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
y_pred
```

```python
new = pd.DataFrame({'Actual':y_test.squeeze(),'Predicted':y_pred.squeeze()})
```

```python
sns.scatterplot(data = new,x ='Actual',y='Predicted')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae = mean_absolute_error(y_test,y_pred)
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
rmse = np.sqrt(mse)
```

```python
print(f'Mean absolute error : {mae:.2f}')
print(f'Mean squared error : {mse:.2f}')
print(f'root Mean absolute error : {rmse:.2f}')
```
